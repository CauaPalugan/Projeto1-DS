```python
import pandas as pd

#Importando Base de Dados

reviews = pd.read_csv("AustralianVehiclePrices.csv")
reviews.columns = ['Marca', 'Ano', 'Modelo', 'Carro/Suv', 'Titulo', 'Condicao', 'Cambio', 'Motor', 'TipoDeDirecao', 
                   'Combustivel', 'Consumo', 'Quilometragem', 'Cor', 'Local', 'Cilindros', 'Tipo', 'Portas', 
                  'Assentos', 'Preco']
reviews.head()
```

```python
#Inspecionando os dados
reviews.Modelo.value_counts().head(10)
```

```python
#Variavel que da vida ao grafico dos 10 carros mais Vendidos
contagemModelos = reviews.Modelo.value_counts().to_frame().reset_index()
contagemModelos.columns = ['modelo', 'total']
contagemModelos.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Grafico de Barras dos 10 carros mais Vendidos

plt.figure(figsize=(10, 6))

sns.barplot(x="modelo", y="total", data = contagemModelos.head(10), palette=sns.color_palette('crest', 10)[::-1])

plt.title("Os 10 Carros mais Vendidos")
plt.xlabel("Modelos")
plt.ylabel("Quantidade")

plt.show()
```

```python
#Tabela que traz somente as informações dos 10 Carros mais vendidos
maisVendidos = reviews.query("Modelo=='Hilux' or Modelo=='Corolla' or Modelo=='Ranger' or Modelo=='Landcruiser' or Modelo=='I30' or Modelo=='RAV4' or Modelo=='Triton' or Modelo=='Commodore' or Modelo=='X-Trail' or Modelo=='Navara'")
maisVendidos.head(19)
```

```python
#Grafico de Marcas sobre os Carros mais vendidos
marcaMaisVendida = maisVendidos.Marca.value_counts().to_frame().reset_index()
marcaMaisVendida.columns = ['marca', 'total']
sns.barplot(x="marca", y="total", data = marcaMaisVendida, palette=sns.color_palette('YlOrBr', 10)[::-1])

plt.title('Marcas dos Carros mais vendidos')
plt.xlabel('Marcas')
plt.ylabel('Quantidade')
```

```python
#Bloco para gerar o Insight de acima de qual ano que as pessoas procuram comprar os carros

#Variavel dos anos dos carros mais vendidos
anoMaisVendido = maisVendidos.Ano.value_counts().to_frame().reset_index()

#Logica para contar quantos carros vendidos tem o ano acima de 2018
quantidadeCarros2018 = anoMaisVendido.query("Ano>=2018")
quantidadeCarros2018.columns = ['modelo', 'quantidade']
soma2018 = quantidadeCarros2018.quantidade.sum()

#logica para contar total geral de carros vendidos
quantidadeGeral = reviews.Modelo.value_counts().to_frame().reset_index().head(10)
quantidadeGeral.columns = ['modelo', 'quantidade']
somaGeral = quantidadeGeral.quantidade.sum()

print(soma2018, somaGeral)
#48,80%
```

```python
#Bloco para ver preferencia dos compradores na questão do cambio
cambioMaisVendido = maisVendidos.Cambio.value_counts()
cambioMaisVendido
#73% Cambio Automatico
```

```python
#Bloco para ver a media dos valores dos carros mais vendidos
maisVendidosPreco = pd.to_numeric(maisVendidos['Preco'], errors='coerce')
maisVendidosPreco.value_counts().to_frame().reset_index()
maisVendidosPreco.median()
```
